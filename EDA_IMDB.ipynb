```python
# Import the required libraries for data analysis and visualization
import pandas as pd 
import numpy as  NP
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
# Load the cleaned IMDb movie dataset into a DataFrame
df=pd.read_csv("IMDB-Movie-Data-clean.csv")
```

```python
# Display the number of rows and columns in the dataset
df.shape 
```

```python
# Display the first five records to understand the dataset structure
df.head()
```

```python
# Generate summary statistics for all numerical columns
df.describe()
```

```python
# Remove the unnecessary index column created during data export
df=df.drop("Unnamed: 0.1",axis=1)   
df=df.drop("Unnamed: 0",axis=1)    
```

```python
df.shape
```

```python
df.describe()
```

```python
# Analyze the average Metascore of movies released each year
plt.figure(figsize=(8,6))
sns.barplot(data=df,x="Year",y="Metascore")
plt.title("Average Metascore by Release Year")
plt.xlabel("Release Year")
plt.ylabel("Average Metascore")
plt.show()
```

```python
# Create pairwise scatter plots and distributions for all numerical features
# to identify relationships, trends, and potential correlations.

sns.pairplot(df)
plt.show()
```

```python
# Calculate the correlation matrix for numerical features
numerical_df = df.select_dtypes(include=['number'])
```

```python

corr_matrix = numerical_df.corr()
```

```python
# Visualize the correlation matrix using a heatmap
# to identify strong positive or negative relationships
# between numerical variables.  
sns.heatmap(corr_matrix,annot=True, fmt='.2f',)
```

```python
# Analyze whether movies with more audience votes tend to generate higher revenue       
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Votes",
    y="Revenue (Millions)"
)

plt.title("Votes vs Revenue")
plt.xlabel("Votes")
plt.ylabel("Revenue (Millions)")
plt.show()
```

```python
# Analyze the relationship between IMDb ratings and the number of audience votes
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Rating",
    y="Votes"
)

plt.title("Rating vs Votes")
plt.xlabel("IMDb Rating")
plt.ylabel("Votes")
plt.show()
```

```python
# Analyze whether movie runtime has any relationship with audience voting
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Runtime (Minutes)",
    y="Votes"
)

plt.title("Runtime vs Votes")
plt.xlabel("Runtime (Minutes)")
plt.ylabel("Votes")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="Votes",
    y="Revenue (Millions)",
    color="royalblue",
    alpha=0.7
)

plt.title("Relationship Between Votes and Revenue", fontsize=15)
plt.xlabel("Votes")
plt.ylabel("Revenue (Millions)")
plt.grid(alpha=0.3)

plt.show()
```
